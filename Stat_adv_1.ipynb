**Q.1 What is a random variable in probability theory?**
- A random variable is a variable whose value is unknown. It is a function that assigns a real value to an outcome of a random variable. It can be either discrete having certain specific values or continuous with any value in a range.
Example: tossing a coin or throwing a dice.

**Q.2 What are the types of random variables?**
- Random variable can be of two following types:
  - *Discrete Random Variable*: A discrete random variable take finite and countable number of values. Example of this is tosssing a coin 2 times, let X (random variable) is number of heads then X can be equal to {0,1,2}. These are the only possible values for random variable and as they are finite and no other value is possible for random variable. So, X is discrete random variable.
  - *Continuous Random Variable*: A continuous random variable take infinite number of values within a specified range or interval. Example: the average height of people in a city or town. It can take any value under a certain range and have no gaps between values. so, we consider it as continuous random variable.

**Q.3 What is the difference between discrete and continuous distributions?**
- Discrete distribution is the distribution in which data can only take certain finite and countable values whereas in continuous distribution data values can take any value within a given specified range. In discrete distributions we use probability mass function to represent the distribution while in continuous distribution we have probability density function. Discrete data typically shows information for a particular event while continuous data shows trends in data over time. Overall, we can say in discrete distributions data can assume specific values whereas continuous distributions can take any numeric value, including fractions and decimals. Examples of discrete distribution is bionomial and poisson and examples of continuous are normal and exponential distributions.

**Q.4 What are probability distribution functions (PDF)?**
- Probability distribution function is a function that gives the probabilty of all possible outcomes of a random variable. In other words, it is a set of all possible outcomes of a random experiment, showing how probabilities are distributed accross the random variable. They are mainly classified as 3 types:
  - *Probability Mass Function (P.M.F)*:
  It assigns a probability to each value of a discrete random variable. It only includes the probabilities of values that are possible. Example: Pmf of bionomial distribution is:
  $$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$
  where, k = 0,1,2....n

  - *Probability Density Function(P.D.F)*:
  Pdf is a probability function for a continuous random variable. It doesn't gives the value of probability directly but probabilities are represented by area under the graphs. So, we can get probility of a variable by integrating over the interval we want to calculate.
  Example of pdf of normal distribution is
  $$f(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \, e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

 - *Cummulative probability function(CDF)*: It is used for both pdf and pmf. It gives the probability of a variable less than or equal to a certain value.
 $$F(x) = P(X \leq x)$$

**Q.5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
- PDF describes the probability distribution of a continuous random variable whereas CDF calculates the cummulative probability of both discrete and continuous random variable. The value of pdf lies between 0 and 1 and the value of cdf is greater than zero. Pdf is probability that a random variable will take a value equal to certain value while cdf is the probability that a random variable will take a value less than or equal to the a certain value.
Pdf formula is:
$$P(a \leq X \leq b) = \int_a^b f(x) \, dx$$
Cdf formula is:
$$F(x)= P(X \leq x)$$

**Q.6 What is a discrete uniform distribution?**

**Q.8 What is the binomial distribution, and how is it used in probability?**
- Bionomial distribution is a discrete distribution having two possible outcomes of each trial. It is same as bernoulli distribution but we have n independent trials and each trial have probability of success as p. Pmf of bionomial distribution is:
  $$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$
  where, k = 0,1,2....n

  It is used whenever we have repeated independent trials in which each trial have only two possible outcomes. Here, we mostly calculate probability of k number of success. Example if we want to caluclate probability of getting three heads in an experiment where a coin is tossed 10 times.

**Q.9 What is the Poisson distribution and where is it applied?**
- Poisson distribution tells us about the probability of number of times a specific event had occured in a given interval of time and space. It only has one parameter known as lambda which is equal average number of occurences happened in that interval. Poisson distribution is a limiting case of the bionomial distribution which arrises when the number of trials increases.
$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad \text{for } k = 0, 1, 2, \dots$$
Poisson distribution is used for counting occurences over continuous intervals. It is mostly useful when we have independent events occuring randomly at a constant rate. Example: no. of accidents occured on a certain road.

**Q.13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
- Central Limit Theorm states that the distribution of the mean of all the samples taken from the population will follow normal distribution, as long as the sample sizes are large enough and donot depend on the shape of original distribution. In this the data must be randomly sampled and the samples must not relate with each other.
CLT is important because it allows us to make inferences about population parameters based on samples. Moreover, it is useful in statistics tests like z-test and t-test because they require normal distributed data. So, even if population is not normally distributed we can calculate the tests taking large samples and calculating there mean.

**Q.16 How do you calculate a Z-score, and what does it represent?**
- Z-scores are calculated by subtracting mean from data point and dividing it with the standard deviation. For a single data point z-score is calculated by:
$$Z = \frac{X - \mu}{\sigma}$$ where,
$\mu$ is mean and $\sigma$ is standard deviation.

  For a sample mean instead of single point z-score is calculated by:
$$Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}$$

  Z-scores represent how many standard deviations a value is from mean. Example if z=2 this mean the data point is 2 standard deviation away from mean. Hence, it helps us to find where a specific data point lies in a distribution.

**Q.22 What is the concept of expected value in a probability distribution?**
- Expected value is the long run average or mean value of a random variable calculated over many trials. It represents the central value of a data.

  For discrete random variable expected value is:
$$E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)$$
where, xi are the outcomes of the random variable

  For continuous random variable expected value is:
  $$E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$$

**Q.23 How does a probability distribution relate to the expected outcome of a random variable?**

Overall, we can say probability distribution gives us the structure and expected value is a summary measure which tells us about the average value we must expect if an experiment is repeated multiple times.
